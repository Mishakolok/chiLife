```python
import numpy as np
import matplotlib.pyplot as plt
import chiLife as xl

omp = xl.fetch('1omp')
SL1 = xl.SpinLabel('R1A', 238, omp, sample=10000)
SL2 = xl.SpinLabel('R1A', 275, omp, sample=10000)

r, Pexp, Plb, Pub = np.loadtxt('Exp_Data.txt')
P = xl.distance_distribution(SL1, SL2, r=r)

xl.save('1omp_S238R1_L275R1.pdb', SL1, SL2, omp)
```

```python
plt.style.use('chiLife')

fig, ax = plt.subplots(figsize=(8, 3))
ax.set_title('MBP S238R1 L275R1')
ax.plot(r, Pexp, label='Experiment')
ax.fill_between(r, Plb, Pub, alpha=0.5)
ax.plot(r, P, label='Prediction')
ax.set_yticks([])
ax.set_xlabel(r'distance ($\rm\AA$)')
ax.legend()
for spine in ['left', 'top', 'right']:
    ax.spines[spine].set_visible(False)

plt.show()
fig.savefig('MBP S238R1 L275R1_Pr.png', bbox_inches='tight', dpi=1000)
```
